```python
# Загрузка данных
data = pd.read_csv("best_selling_books_2.csv")

# Первичный анализ структуры данных
print("Форма данных:", data.shape) # Показывает размер таблицы - количество строк (образцов данных) и столбцов (признаков).
print("\nТипы данных:")
print(data.dtypes) # Типы данных для каждого столбца
print("\nОбзор первых строк:")
print(data.head()) # Показывает первые 5 строк таблицы.
```

```python
# Проверка на пропущенные значения
print("\nПропущенные значения:")
print(data.isnull().sum()) # Проверяет каждый столбец на наличие пропущенных значений (NaN).

# Основные статистические показатели
print("\nСтатистика данных:")
print(data.describe()) # Выводит базовую статистику числовых данных. Помогает понять распределение данных и их диапазоны
```

Столбец Volume Sales(объем продаж) нужно перевести в тип данных с плавающей точкой, чтобы потом нормально строились граффики

```python
# Удалим лишние символы, оставив только цифры и точку
data["Volume Sales"] = data["Volume Sales"].replace({r'[^\d.]': ''}, regex=True)

# Преобразуем столбец в тип float
data["Volume Sales"] = pd.to_numeric(data["Volume Sales"], errors="coerce")

# Проверим результат
print("\nТипы данных после преобразования:")
print(data.dtypes)

# Посмотрим на первые строки для проверки
print("\nПервые строки данных после преобразования Volume Sales в float:")
print(data.head())

# Проверим, есть ли пропущенные значения
print("\nКоличество пропущенных значений в Volume Sales:", data["Volume Sales"].isnull().sum())
```

```python
import seaborn as sns #библиотека для создания статистических графиков
import matplotlib.pyplot as plt #библиотека для визуализации данных

# График распределения объёмов продаж
plt.figure(figsize=(10, 6))
sns.histplot(data["Volume Sales"], bins=20, kde=True)
plt.title("Распределение объёмов продаж")
plt.xlabel("Объем продаж (в млн)")
plt.ylabel("Частота")
plt.show()

# Топ-10 авторов по объему продаж
top_authors = data.groupby("Author")["Volume Sales"].sum().nlargest(10)
plt.figure(figsize=(12, 8))
top_authors.plot(kind='bar', color='skyblue')
plt.title("Топ-10 авторов по объёму продаж")
plt.xlabel("Автор")
plt.ylabel("Объем продаж (в млн)")
plt.xticks(rotation=45)
plt.show()

# Распределение по жанрам
plt.figure(figsize=(10, 6))
sns.countplot(y="Genre", data=data, order=data["Genre"].value_counts().index, palette="viridis")
plt.title("Распределение книг по жанрам")
plt.xlabel("Количество")
plt.ylabel("Жанр")
plt.show()

# Влияние издателей на объем продаж
top_publishers = data.groupby("Publisher")["Volume Sales"].sum().nlargest(10)
plt.figure(figsize=(12, 8))
top_publishers.plot(kind='bar', color='orange')
plt.title("Топ-10 издателей по объёмам продаж")
plt.xlabel("Издатель")
plt.ylabel("Объем продаж (в млн)")
plt.xticks(rotation=45)
plt.show()
```
