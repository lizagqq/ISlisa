Индивидуальное задание

```python
data = pd.read_csv('bodyfat.csv')
data.head()
```

```python
import statsmodels.api as sm
import pandas as pd

# Загрузка и подготовка данных
#  X - признаки, y - целевая переменная (BodyFat)
X = data.drop('BodyFat', axis=1)  # Признаки
y = data['BodyFat']  # Целевая переменная
```

```python
# Добавляем столбец единиц для свободного члена
X = sm.add_constant(X)

# Строим модель линейной регрессии
model = sm.OLS(y, X).fit()

# Выводим результаты модели
print(model.summary())

# Применяем Backward Elimination
# Задаем порог для p-уровня
p_threshold = 0.05
```

```python
# Процесс исключения признаков
while True:
    # Находим наибольший p-уровень среди признаков
    p_values = model.pvalues
    max_p_value = p_values.max()

    # Если наибольший p-уровень больше порога, удаляем признак с этим p-уровнем
    if max_p_value > p_threshold:
        feature_with_max_p = p_values.idxmax()  # Признак с максимальным p-уровнем
        X = X.drop(columns=[feature_with_max_p])  # Удаляем его из данных
        model = sm.OLS(y, X).fit()  # Строим модель с оставшимися признаками
    else:
        break
```

```python
# Результаты финальной модели
print(model.summary())
```
